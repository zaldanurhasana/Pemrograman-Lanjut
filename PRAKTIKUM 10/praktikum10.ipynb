```python
mahasiswa = ["Zalda", "Saci", "Pipi", "Putri", "Kolbi", "Zaza"]
mahasiswa.append("Dika")
mahasiswa.insert(3, "Rehan")
mahasiswa.pop()
mahasiswa.sort(reverse=True)
mahasiswa.remove("Kolbi")

i = 1
for nama in mahasiswa:
    print("Nama Mahasiswa ke-", i ,": ", nama)
    i+=1
    
print(mahasiswa)
```

```python
person = {"name": "Zaza", "hp": "080808", "kelas": "TEKOM F"}
print(person)
```

### Analisis
Kode ini membuat sebuah dictionary bernama person yang berisi tiga pasangan key dan value: "name": "Zaza", "hp": "080808", dan "kelas": "TEKOM F". Ketika perintah print(person) dijalankan, program akan menampilkan seluruh isi dictionary tersebut di layar dalam format {key: value}.

```python
person = {"name": "Zaza", "hp": "080808", "kelas": "TEKOM F"}
for key, value in person.items():
    print(key, ":", value)
```

### Analisis

```python
i = 1
while i <= 9:
    print(i)
    i += 2
```

### Analisis
Kode ini menggunakan perulangan while untuk menampilkan angka dari 1 hingga 9 dengan penambahan 2 di setiap langkah. Variabel i dimulai dari 1 dan terus bertambah 2 setiap kali perulangan dijalankan, sehingga angka yang dicetak berturut-turut adalah 1, 3, 5, 7, dan 9. Perulangan akan berhenti ketika nilai i sudah lebih besar dari 9.

```python
i = 1
while i <= 13:
    if i == 7:
        print("WOW")
    else:
        print(i)
    i +=2
```

### Analisis
Kode ini memakai perulangan while yang menambah nilai i sebesar 2 pada setiap iterasi, dimulai dari 1 hingga 13. Ketika i bernilai 7, program menampilkan teks "WOW", sedangkan untuk nilai lainnya, program mencetak angka seperti biasa. Hasil akhirnya adalah 1, 3, 5, WOW, 9, 11, dan 13.

```python
for i in range(1, 11):
    if i <= 8:
        if i % 2 == 1:
            print("hidup")
        else:
            print("mahasiswa")
    else:
        print(i)
```

### Analisis
Kode ini menjalankan perulangan for dari angka 1 sampai 10. Di dalamnya ada percabangan: jika i kurang dari atau sama dengan 8, program akan memeriksa apakah i termasuk bilangan ganjil atau genap. Jika ganjil (i % 2 == 1), maka program mencetak "hidup", dan jika genap, mencetak "mahasiswa". Untuk nilai i di atas 8 (yaitu 9 dan 10), program langsung menampilkan angkanya. Jadi, hasil outputnya berupa kombinasi antara kata dan angka sesuai kondisi tersebut.

```python
for i in range(1, 14):
    print(i)
    if i % 3 == 2:  
        print("Ha")
```

### Analisis
Kode ini menjalankan perulangan for dari angka 1 sampai 13. Setiap nilai i dicetak lebih dulu, lalu dicek apakah sisa bagi i dengan 3 sama dengan 2 (i % 3 == 2). Jika kondisi itu benar, program akan menampilkan kata "Ha". Jadi, setiap kali angka memiliki sisa 2 saat dibagi 3, program mencetak "Ha" setelah angka tersebut. Hasil akhirnya berupa kombinasi angka dan kata "Ha" sesuai aturan yang berlaku.

```python
for i in range(1, 14):
    if i not in [3, 5, 6, 9, 10, 12]:
        print(i)
        
    if i in [2, 5, 8, 11]:
        print("Ha")

    if i in [4, 9]:
        print("Hi")
```

### Analisis
Kode ini menjalankan perulangan dari 1 hingga 13 dan menampilkan angka atau kata tertentu sesuai kondisi yang diberikan. Jika angka tidak ada dalam daftar [3, 5, 6, 9, 10, 12], maka program mencetak angkanya. Jika angka termasuk dalam [2, 5, 8, 11], program menampilkan "Ha", dan jika termasuk dalam [4, 9], menampilkan "Hi". Hasil akhirnya berupa gabungan antara angka dan kata sesuai dengan aturan tersebut.
